```python
%%sql players_info <<
SELECT 
    player_info.name, 
    player_info.account, 
    bosses.boss, 
    profession.professions
FROM player_info JOIN bosses JOIN profession ON player_info.boss_id = bosses.id AND player_info.profession_id = profession.id
WHERE name NOT LIKE "Conjured%" AND name NOT LIKE "Saul%";

```
